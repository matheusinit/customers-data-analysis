#### Clientes

```python
customers_columns = ['Nome/Razão Social', 'Nome Fantasia', 'CPF/CNPJ', 'Data Nasc.', 'Data Cadastro cliente']

customers = data[customers_columns]

value_counts = customers['CPF/CNPJ'].value_counts()
print(value_counts)

customers.head()
```

Planos

```python
plan_column = data['Plano']
plan_cost_column = data['Plano Valor']

plan_selected_columns = ['Plano', 'Plano Valor']
plan_unique_values = data[plan_selected_columns].drop_duplicates()

plan = plan_unique_values

value_counts = plan.value_counts()
print(value_counts)

plan.info()
```

Status do contrato

```python
from pandas import DataFrame

contract_statuses_column = data['Status']

contract_statuses_column = contract_statuses_column.drop_duplicates()

contract_statuses_column = contract_statuses_column.reset_index(drop=True)

contract_statuses = DataFrame({
  'status': contract_statuses_column
})

contract_statuses.value_counts()
```
